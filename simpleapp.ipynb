```python
import langchain
```

Loading from Text, PDF, Link

```python
# Loading from a text file
from langchain_community.document_loaders import TextLoader
text_value = TextLoader('./speech.txt').load()
print(text_value)
```

```python
# Loading from PDF File
from langchain_community.document_loaders import PyPDFLoader
pdf_value = PyPDFLoader('./attention.pdf').load()
print(pdf_value)
```

```python
type(pdf_value[0]) # document type
```

```python
#Loading from Web Page
from langchain_community.document_loaders import WebBaseLoader
web_value = WebBaseLoader(web_path='https://mohammedfahimullah.medium.com/go-pro-with-rxjs-2f868e84cb3f').load()
print(web_value)
```

```python
print(web_value[0].metadata)
```

Split by Characters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500,chunk_overlap = 50,separators='\n')
splitted_values = text_splitter.split_documents(pdf_value)
print(splitted_values)
print(len(splitted_values))
```

```python
final_documents_list = [data.page_content for data in splitted_values]
print(len(final_documents_list))
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

model = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
embeddings = [model.embed_query(embedding) for embedding in final_documents_list]
print(embeddings)
```

```python
print(len(embeddings))
```

Vector store db

```python
from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(documents=splitted_values,embedding=model,persist_directory='./chroma-db')
```

Similarity Search

```python
query = 'What is my future ?'
retrieved_results = db.similarity_search_with_score(query=query)
print(retrieved_results)
```

```python
print(len(retrieved_results))
```

```python
for document,score in retrieved_results:
    print(document.page_content, score)
```
